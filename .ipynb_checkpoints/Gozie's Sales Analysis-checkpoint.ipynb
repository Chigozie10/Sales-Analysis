# Sales Analysis

### Import necessary libraries

```python
import pandas as pd
import os
```

#### Task #1 = Merge the 12 months of sales data into a single csv file

```python
df = pd.read_csv("./Sales_Data/Sales_April_2019.csv")

# getting the list of all files in the Sales Data folder
files = [file for file in os.listdir('./Sales_Data')]

# creating an empty dataframe to store all the data files
all_months_data = pd.DataFrame()

# merge all the months data file into one file
for file in files:
    df = pd.read_csv("./Sales_Data/"+file)
    all_months_data = pd.concat([all_months_data, df])


# writting your data to a csv file    
all_months_data.to_csv("all_data.csv", index=False)

# for file in files:
#     print(file)
```

#### Read in updated dataframe

```python
all_data = pd.read_csv("all_data.csv")
all_data.head()
```

### Data Cleaning 

##### Drop rows of NAN

```python
# Find all the NAN values
nan_df = all_data[all_data.isna().any(axis = 1)]
nan_df.head()

all_data = all_data.dropna(how="all")
all_data.head()
```

#### Find 'Or' and delete it

```python
temp_df = all_data[all_data['Order Date'].str[0:2] != 'Or']
all_data = temp_df
all_data.head()
```

##### convert columns to the correct data type

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered']) # Make int
all_data['Price Each'] = pd.to_numeric(all_data['Price Each']) # Make float
all_data.head()
```

### Augment data with additional columns

#### Task 2: Adding the month column to the merged data

```python
# This adds a new column to the dataframe

# dropping columns
# all_data.drop(columns='Column', inplace=True)


# all_data = all_data.drop(columns=['month','Month'])

# add month column
all_data['month'] = 3
all_data.head()
# drop 'month' column
all_data = all_data.drop(['month'], axis=1)
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

#### Task 3: Add a sales column

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()

# to select rows whose column values are a value, basically filtering for some rows
# df1 = all_data.loc[all_data['Quantity Ordered'] == 2]
```

#### Task 4: Add a city column

```python
# let's use the .apply() method

# explanation of the code: the x value is the cell content of the column purchase address; for each of the cell 
# content split it by commas and then grab the first index.
# all_data['City'] = all_data['Purchase Address'].apply(lambda x: x.split(',')[1])

# another way to do it is using a defined function

def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

# all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x) + ' (' + get_state(x) + ') ')

# using f strings to look better
all_data['City'] = all_data['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_state(x)})")

all_data.head()
```

```python
all_sum = all_data.groupby('Month').sum()
all_sum.head()

# to just see the summed up sales column by month
# all_data.groupby('Month').sum()['Sales']
```

```python
import matplotlib.pyplot as plt

months = range(1,13)

plt.bar(months, all_sum['Sales'])
plt.xticks(months)
plt.ylabel('Total sales in USD ($)')
plt.xlabel('Months')
plt.title('Total sales by month')
plt.show()
```

#### Question 3: What time should we display advertisements to maximize likelihood of customer's buying product?

```python

all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
all_data.head()
```

```python
all_data['Hour'] = all_data['Order Date'].dt.hour

all_data.head()
```

```python
hours = [hour for hour, df in all_data.groupby('Hour')]

print(hours)

plt.plot(hours, all_data.groupby('Hour').count()['Order ID'])
plt.xticks(hours)
plt.xlabel('Hour')
plt.ylabel('Number of Orders')
plt.grid()
plt.show()
all_data.groupby('Hour').count()

# My recommendation is around 11am and 7pm.
```

#### Question 4: What products are most often sold together?

```python
df = all_data[all_data['Order ID'].duplicated(keep = False)]

df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

df.head()

df = df[['Order ID','Grouped']].drop_duplicates()

df.head()
```

```python
# Now let's count up the pairs that appears most frequently

from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    rowlist = row.split(',')
    count.update(Counter(combinations(rowlist, 2)))
    
for key, value in count.most_common(10):
    print(key, value)
```

#### Question 5: What products sold the most ? and why did it sell the most ?

```python
all_data.head()

product_group = all_data.groupby('Product')

quantity_ordered = product_group.sum()['Quantity Ordered']

print(quantity_ordered)

products = [product for product, df in product_group]

plt.bar(products, quantity_ordered)
plt.xticks(products, rotation = 'vertical', size = 8)
plt.ylabel('Quantity Ordered')
plt.xlabel('Products')
plt.title('Quantity ordered by products')
plt.show()


```

```python
# Hypothesis: AAA Batteries are very cheap that's why it sold the most and LG Dryers/washing machines are expensive 
# hence selling the least

# As a data scientist, we got to prove our hypothesis.

prices = all_data.groupby('Product').mean()['Price Each']

print(prices)

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color ='g')
ax2.plot(products, prices, 'b-')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color ='g')
ax2.set_ylabel('Price', color ='b')
ax1.set_xticklabels(products, rotation = 'vertical', size = 8)

plt.show()
```
